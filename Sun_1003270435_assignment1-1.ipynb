2. [4pts] Estimating the difference between average salary (Q24) of men vs. women (Q2).

a. [0.5pts] Compute and report descriptive statistics for each group (remove missing data, if
necessary).

b. [0.5pts] If suitable, perform a two-sample t-test with a 0.05 threshold. Explain your rationale.

c. [1.5pts] Bootstrap your data for comparing the mean of salary (Q24) for the two groups. Note that the number of instances you sample from each group should be relative to its size. Use 1000 replications. Plot two bootstrapped distributions (for men and women) and the distribution of the difference in means.

d. [0.5pts] If suitable, perform a two-sample t-test with a 0.05 threshold on the bootstrapped data. Explain your rationale.

e. [1pts] Comment on your findings.


```python
#a
Woman_2a = kaggle.loc[kaggle.Q2=='Woman']
salary_woman_2a = Woman_2a.loc[:,['Q2','Q24']]
print(salary_woman_2a.isnull().sum())
salary_woman_2a.describe()
```

```python
Man_2a = kaggle.loc[kaggle.Q2=='Man']
Man_2a
salary_man_2a = Man_2a.loc[:,['Q2','Q24']]
print(salary_man_2a.isnull().sum())
salary_man_2a.describe()
```

```python
#b
from scipy import stats
print(stats.ttest_ind(salary_man_2a['Q24'], salary_woman_2a['Q24'],equal_var = False))
print('According to the ttest, the Pvalue is smaller to 0.05, so it does not have equal mean, we need to reject the null hypothesis.')
```

According to the pvalue, because the pvalue is near 0, so the two group has significantly different, so we can conclude that mean of salary of man is not equal (greater) than that of woman.

```python
#c
import seaborn as sns
import random
male = kaggle.loc[kaggle.Q2=='Man']['Q24']
female = kaggle.loc[kaggle.Q2=='Woman']['Q24']
def boot(sample):
    salary_series = random.choices(sample,k=len(sample))
    array_mean = np.array(salary_series).mean()
    return array_mean

men_bootstrap=[]
for i in range(1000):
    x=boot(male.tolist())
    men_bootstrap.append(x)
women_bootstrap=[]
for i in range(1000):
    x=boot(female.tolist())
    women_bootstrap.append(x)
    

#plt.hist(women_bootstrap, kde = True, color = 'pink')
plt.hist(men_bootstrap, color = 'blue')
plt.xlabel('Salary')

plt.hist(women_bootstrap, color = 'pink')
plt.xlabel('Salary')
plt.legend(labels=['men_bootstrap','women_bootstrap'])
```

```python
#d
stats.ttest_ind(men_bootstrap,women_bootstrap,equal_var = False)
```

According to the pvalue, because the pvalue is 0, so the two group has significantly different, so we can conclude that mean of salary of man is not equal (greater) than that of woman.

```python
#e
print('Both the mean of salary between Man and Woman P-value are near 0, so we can know that the mean salary of man is greater than the mean salary of Woman.')
```

[5pts] Select “highest level of formal education” (Q4) from the dataset and repeat steps a to e, this time use analysis of variance (ANOVA) instead of t-test for hypothesis testing to compare the means of salary for three groups (Bachelor’s degree, Doctoral degree, and Master’s degree) [0.75pts for a; 0.5 pts for b; 2pts for c; 0.75 pts for d; 1pt for e].

```python
#a

Bachelor_3a_salary = kaggle.loc[kaggle.Q4=='Bachelor’s degree']['Q24']
Master_3a_salary=kaggle.loc[kaggle.Q4=='Master’s degree']['Q24']
Doctoral_3a_salary=kaggle.loc[kaggle.Q4=='Doctoral degree']['Q24']


print(Bachelor_3a_salary.isnull().sum())
print(Master_3a_salary.isnull().sum())
print(Doctoral_3a_salary.isnull().sum())

Bachelor_describe=pd.DataFrame(Bachelor_3a_salary).describe().rename_axis('Summary for Bachelor')
Bachelor_describe
```

```python
Master_describe=pd.DataFrame(Master_3a_salary).describe().rename_axis('Summary for Master')
Master_describe
```

```python
Doctoral_describe=pd.DataFrame(Doctoral_3a_salary).describe().rename_axis('Summary for Doctoral')
Doctoral_describe
```

```python
#b
import scipy.stats as stats
f_value, p_value = stats.f_oneway(kaggle.loc[kaggle.Q4=='Bachelor’s degree']['Q24'], kaggle.loc[kaggle.Q4=='Master’s degree']['Q24'], kaggle.loc[kaggle.Q4=='Doctoral degree']['Q24'])
print("F_value is",f_value)
print("P_value is",p_value)

```

```python
#c
import seaborn as sns
Bachelor_boot = kaggle.loc[kaggle.Q4=='Bachelor’s degree']['Q24']
Master_boot = kaggle.loc[kaggle.Q4=='Master’s degree']['Q24']
Doctoral_boot = kaggle.loc[kaggle.Q4=='Doctoral degree']['Q24']

def education_boot(sample_education):
    salary_series_education = random.choices(sample_education,k=len(sample_education))
    array_mean_education = np.array(salary_series_education).mean()
    return array_mean_education

Bachelor_bootstrap=[]
for i in range(1000):
    x=education_boot(Bachelor_boot.tolist())
    Bachelor_bootstrap.append(x)
Master_bootstrap=[]
for i in range(1000):
    x=education_boot(Master_boot.tolist())
    Master_bootstrap.append(x)
Doctoral_bootstrap=[]
for i in range(1000):
    x=education_boot(Doctoral_boot.tolist())
    Doctoral_bootstrap.append(x)    

sns.histplot(data = Bachelor_bootstrap, kde = True, color='Blue')
sns.histplot(data = Master_bootstrap, kde = True, color='Orange')
sns.histplot(data = Doctoral_bootstrap, kde = True, color='gray')
plt.xlabel('Salary')
plt.legend(labels=['Bachelor_bootstrap', 'Master_bootstrap', 'Doctoral_bootstrap'])
```

```python
#d
import scipy.stats as stats
f_value, p_value = stats.f_oneway(Bachelor_bootstrap, Master_bootstrap, Doctoral_bootstrap)
print("F_value is",f_value)
print("P_value is",p_value)
print('The pvalue is 0, so we can conclude that the three group (Bachelor, Master, Doctoral) mean are significantly different, so we can say that the mean of salary for three groups are different from each other')

```
